```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as ss
from IPython.display import Math, Latex
import warnings
warnings.filterwarnings("ignore")
```

# Continuous Uniform Distribution

```python
n= 10000
start= 10
end=20
uniform_data= ss.uniform.rvs(size=n,loc=start,scale= end)
```

```python
ax=sns.distplot(uniform_data,
             bins=20,
            kde=True,
            )
ax.set(xlabel="Uniform Distribution",ylabel="Frequency")
```

# Normal Distribution

```python
normal_data= ss.norm.rvs(size= 5000,loc=5,scale=20)
```

```python
ax=sns.distplot(normal_data,
             bins=20,
            kde=True,
            )
ax.set(xlabel="Normal Distribution",ylabel="Frequency")
```

# Exponential Distribution

```python
exp_data=ss.expon.rvs(size=1000,loc=5,scale=50)
ax=sns.distplot(exp_data,
             bins=20,
            kde=True,
            )
ax.set(xlabel="Exponential Distribution",ylabel="Frequency")
```

# Chi-Square Distribution

```python
chi_data=ss.chi2.rvs(df=55,size=1000)
```

```python
ax=sns.distplot(chi_data,
             bins=20,
            kde=True,
            )
ax.set(xlabel="Chi Square Distribution",ylabel="Frequency")
```

# Weibull Distribution

```python
import matplotlib.pyplot as plt
x = np.arange(1,100.)/50.
def weib(x,n,a):
    return (a / n) * (x / n)**(a - 1) * np.exp(-(x / n)**a)

count, bins, ignored = plt.hist(np.random.weibull(5.,1000))
x = np.arange(1,100.)/50.
scale = count.max()/weib(x, 1., 5.).max()
plt.plot(x, weib(x, 1., 5.)*scale)
plt.show()
```

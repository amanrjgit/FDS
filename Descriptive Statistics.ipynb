```python
import pandas as pd
```

```python
df= pd.DataFrame({"Name":["Jay","Meet","Zac","Roy","Amar","Aman","Jeet","Sam","Jam","Sunil"],
                  "Salary":[2000,3078,4659,6549,7653,8943,7354,6587,3267,7653],
                  "Country":["IND","ENG","IND","IND","ARG","IND","BHU","IND","AUS","USA"]})
df
```

```python
print("Mean: ",df.Salary.mean())
print("Median: ",df.Salary.median())
print("Mode: ",df.Salary.mode()[0])
```

```python
print("Variance: ",df.Salary.var())
print("Standard Deviation: ",df.Salary.std())
```

```python
print("Maximum Salary: ",df.Salary.max())
print("Minimum Salary: ",df.Salary.min())
```

```python
df.groupby("Country")["Salary"].sum().reset_index(name="Totol_salary")
```

```python
print("Skewness: ",df.Salary.skew())
print("Kurtosis: ",df.Salary.kurtosis())
```

# Problem Statement


### Randomly selected patient has the test and it comes back positive, what is the probability that the patient has cancer? In our scenario we were given 3 pieces of information, the the base rate, the sensitivity (or true positive rate), and the specificity (or true negative rate).<br>
 
Sensitivity: 85% of people with cancer will get a positive test result<br> 
Base Rate: 0.02% of people have cancer. <br>
Specificity: 95% of people without cancer will get a negative test result<br>

```python
# Let A be the event a person has cancer
# LET B be the event that test is positive 
PcancerT=0.0002
PtestF_cancerF=0.95
PtestT_cancerT=0.85

#To Find: PcancerT_testT
#95% of people without cancer will get a negative test result
#i.e there is a 5% chance that people without cancer will get a positive result
```

### To Find: PcancerT_testT
PcancerT_testT= PtestT_cancerT*PcancerT/PtestT

We need to fine the value for PtestT
PtestT= PtestT_cancerT*PcancerT + PtestT_cancerF*PcancerF

To find the PtestT we need PtestT_cancerF
PtestT_cancerF=1-PtestF_cancerF

```python
PtestT_cancerF=1-PtestF_cancerF
PtestT_cancerF
```

```python
#Hence, 
PcancerF=1-PcancerT
print("PcancerF : ",PcancerF)
print(" ")
PtestT= PtestT_cancerT*PcancerT + PtestT_cancerF*PcancerF
print("PtestT : ",PtestT)
```

```python
PcancerT_testT=(PtestT_cancerT*PcancerT)/PtestT
print("The probability that the patient has cancer given that the test is positive",round(PcancerT_testT*100,2),"%")
```

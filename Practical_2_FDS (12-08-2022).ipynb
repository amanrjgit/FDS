```python
plt.figure(figsize=(12, 7))
 
countries=["India", "USA", "Brazil", "Russia", "Spain", "UK"]
total_cases=[20000, 40000, 8000, 50000, 4000, 10000]
total_deaths=[1880, 37312, 5971, 45789, 3488, 9008]

 
for i in range(len(countries)):
    plt.bar(countries[i], total_deaths[i], bottom=total_cases[i]-total_deaths[i], color="black")
    plt.bar(countries[i], total_cases[i]-total_deaths[i], color="red")
```

```python
data={
    "countries":["India", "USA", "Brazil", "Russia", "Spain", "UK"],
    "total_cases":[20000, 40000, 8000, 50000, 4000, 10000],
    "total_deaths":[1880, 37312, 5971, 45789, 3488, 9008]
}

df2= pd.DataFrame(data)
df2.plot(x="countries",y=["total_cases","total_deaths"],kind="bar")
```

```python
X_axis = np.arange(len(countries))
  
plt.bar(X_axis - 0.2, total_cases, 0.4, label = 'Total cases',color="red")
plt.bar(X_axis + 0.2, total_deaths, 0.4, label = 'Total deaths',color="grey")
  
plt.xticks(X_axis, countries)
plt.xlabel("Countries")
plt.ylabel("Number of people")
plt.title("Deaths and cases")
plt.legend()
plt.show()
```
